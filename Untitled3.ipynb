```python
import numpy as np
import pandas as pd
import math
import random
```

```python
df=pd.read_csv("25.csv")
df.groupby(['Internships','CGPA']).count()
```

```python
df['DSA'] = np.nan
```

```python
def mlf(x,y,cc,paisa1,paisa2,z,dsa):
    tppr =df[(df['CGPA']==x) & (df['Internships']==y)]
    tppr_index=tppr.index.tolist()
    df.loc[tppr_index]
    df['Codechef'].loc[tppr_index]=cc
    df['Salary'].loc[tppr_index]=random.sample(range(paisa1,paisa2),z)
    df['DSA'].loc[tppr_index]=dsa
```

```python
mlf(5,0,2,10000,15000,49,100)
```
